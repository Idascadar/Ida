```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/content/birds.jpg")
```

```python
type(image)
```

```python
plt.imshow(image)
```

```python
new_image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(new_image)
```

```python
"""
1.spliting image channels
"""
r,g,b =cv2.split(new_image)
print('r', r.shape)
print('g', g.shape)
print('b', b.shape)

```

```python
"""
merge a image
"""
r,g,b =cv2.split(new_image)
new_image = cv2.merge((r,g,b))
```

```python
"""
Resize a image
"""
s = 10
w = int(new_image.shape[1]*s/100)
h = int(new_image.shape[1]*s/100)
dim = (w,h)
re_size = cv2.resize(new_image, dim, interpolation = cv2.INTER_AREA)
re_size.shape
```

```python
"""
3.Rotate operation
"""
(h, w) = new_image.shape[:2]

c = (w/2, h/2)

angle = 90

m = cv2.getRotationMatrix2D(c, angle, 1.0)
rotate_90 = cv2.warpAffine(new_image, m, (h,w))
```

```python
plt.imshow(rotate_90)
```
